```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import BaggingClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
data
```

```python
# splitting data into two parts input nd target
X = data.data
y = data.target
```

```python
# splitting data into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
# create a model of KNeighboursClassifier
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
knn.score(X_test, y_test)
```

```python
y_pred = knn.predict(X_test)
y_pred
```

```python
# create a model of baggingClassifier
bag_knn = BaggingClassifier(knn, n_estimators = 10)
bag_knn.fit(X_train, y_train)
bag_knn.score(X_test, y_test)
```

```python
y_pred_bag_knn = bag_knn.predict(X_test)
y_pred_bag_knn
```
