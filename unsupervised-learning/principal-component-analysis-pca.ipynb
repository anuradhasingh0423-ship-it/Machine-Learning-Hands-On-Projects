```python
from sklearn.preprocessing import StandardScaler
import pandas as pd

scaler = StandardScaler()
data_scaled_np = scaler.fit_transform(data)   # NumPy

dataframe = pd.DataFrame(
    data_scaled_np,
    columns=['x1', 'x2']
)

dataframe.describe()
```

1. To check whether the data is standard normal form or not. As our data is in standard normal form, we converting our data to standard normal form.

2. Covariance Matrix between the two features that we have (x1,x2)

```python
## Approach 1
c1 = dataframe.x1
c2 = dataframe.x2
np.cov(c1, c2)
```

```python
## Approach 2
covarianceMatrix = dataframe.T @ dataframe / 5
covarianceMatrix
```

```python
## Approach3
np.sum(c1 *c2)/5
```

3. Evaluate Eigen values and eigen vectors of the above correlation matrix

```python
eigenValues, eigenVectors = np.linalg.eig(covarianceMatrix)
```

```python
## eignValues = features
## eigenValues represents the strength of the information given by the eigenvector
eigenValues
```

```python
eigenVectors
```

```python
eigenVectors[:, 1]
# here : means consider all the rows and 1 means give me 1st column
```

```python
dataframe
```

```python
## PC1 - contains the maximum information of the original two that you have
PC1 = dataframe @ eigenVectors[:, 0]
PC1
```

```python
## PC2
PC2 = dataframe @ eigenVectors[:, 1]
PC2
```
